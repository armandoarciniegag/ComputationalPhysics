## Subgráficas

La función subplot permite subdividir una ventana de figuras en varias celdas tal que sea posible realizar una representación gráfica distinta en cada una de ellas

Sintaxis: subplot (m,n,k), donde m,n y k enteros. La ventana de figura será subdividida en mxn celdas, k será el lugar donde pondremos la gráfica. Siempre k>m*n, suponiendo que nos desplazamos por las columnas.

```python
from pylab import *
def f(t):
    return cos(2*pi*t)*exp(-t)
t1=arange(0.0,5.0,0.1)
t2=arange(0.0,5.0,0.02)

subplot(211) #2 renglones, 1 columna, 1 primer gráfica (renglon,columna,grafica)
plot(t1,f(t1),'bo',t2,f(t2),'k')
grid(True)
title('Dos subgraficas')
ylabel('Amortiguamiento')

subplot(212)  #212 2 renglones 1 columna 2 segunda grafica
plot(t2,cos(2*pi*t2),'r>')
grid(True)
xlabel('tiempo (s)')
ylabel('sin amortiguamiento')
show()
```

$$c: I \in \mathbb{R} \rightarrow \mathbb{R}^2$$

###  2. Hipocicloides 

Para $R=1$ y $k=\frac{R_1}{R}$ y tratar de [$-8\pi$ a $8\pi$]

\begin{eqnarray}
x=(R_1-R)cos(t)+R cos\left(\frac{R_1-R}{R}t\right)\\
x=(R_1-R)sin(t)-R sin\left(\frac{R_1-R}{R}t\right)\\
\end{eqnarray}

```python
from pylab import *
k=3.9
R=1 #Se puede ir cambiabo para obtener los distintos hipocicloides
#R1=k *R

theta=arange(-8*pi,8*pi,0.01)
plot((k-R)*cos(theta) + R*cos((k-R/R)*theta)      ,     (k-R)*sin(theta) - R*sin((k-R/R)*theta))
grid()
xlim(-5,5)
ylim(-4,4)
    
show() #Con esto se muestra la gráfica
```

### Histogramas 

Para graficar valores discretos o distribuciones estadísticas es útil usar un histograma. También son útiles para graficar una distribución de números que cruzan un rango de posibles valores. Ejemplo: Tomar una lista de números aleatorios y ver cuantas veces de ocurrencia que presentan. Los números aleatorios serán generados con  distribuciones gaussianas.

```python
import matplotlib.pyplot as p
from random import *   #random da todas las funciones de cosas aleatorias
numeros = normal(size=1000)  #distribucion
p.hist(numeros   #plot histograma
p.title("Histograma Gaussiano")
p.xlabel("Valor")
p.ylabel("Frecuencia")
p.show()
```

1. Nos referimos a "bins"  como el número de distintas categorías. El default en python es 10 bins. 
2. Si queremos cambiar la resolución de nuestra gráfica, es decir, cambiar el número de bins a 20 usamos p.hist(numeros, bins=20)
3. Si queremos graficar una distribución de probabilidad en lugar de un conteo de frecuencias, se tendrá que dividir entre una constante de normalización. Por lo general, se quiere que el 100% de probabilidad se encuentre en 1. p.hist(numeros, bins=20,normed=True) 
4. Si queremos cambiar la apariencia del histograma, que los cuadros no estén llenados, sino que sólo se vean los escalones usamos la opción histtype p.hist(numeros, bins=20,normed=True,histtype='step'

### Ejemplo
Hacer un histograma con la distribución gaussiana y una distribución uniforme de números aleatorios desde -3 a 3 en rojo con 50% de transparencia sobre el color azúl de la distribución gaussiana: 

```python
import matplotlib.pyplot as plt
from numpy.random import normal, uniform
numeros = normal(size=1000)
uniformes = uniform(low=-3, high=3, size=1000)
plt.hist(numeros, bins=20, histtype='stepfilled', normed=True, color='b',label='Gaussiana')
plt.hist(uniformes, bins=20, histtype='stepfilled', normed=True, color='r', alpha=0.5, label='Uniforme')
plt.title("Histograma Gaussiana/Uniforme")
plt.xlabel("Valor")
plt.ylabel("Probabilidad")
plt.legend()
plt.savefig('uniforme.jpeg')
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

mpl.rcParams['legend.fontsize'] = 10  #tamaño de letra

fig = plt.figure()  #se define "fig" a la figura
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)   #linspace hace una escala uniforme, genera una lista uniforme del rango que uno define (-a,a,p) p el paso
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='helicoidal')
ax.legend()

plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Generar el mallado del toro
angle = np.linspace(0, 2 * np.pi, 32)
theta, phi = np.meshgrid(angle, angle) # meshgrid hace el mallado gris de fondo
r, R = .25, 1.
X = (R + r * np.cos(phi)) * np.cos(theta)
Y = (R + r * np.cos(phi)) * np.sin(theta)
Z = r * np.sin(phi)

# desplegar mallado
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)
ax.plot_surface(X, Y, Z, color = 'w', rstride = 1, cstride = 1)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import colors

fig = plt.figure()
ax = fig.gca(projection='3d')

N = 100
y = np.ones((N,1))
x = np.arange(1,N + 1)
z = 5*np.sin(x/5.)

cn = colors.Normalize(min(z), max(z)) # creates a Normalize object for these z values
for i in range(N-1):
    ax.plot(x[i:i+2], y[i:i+2], z[i:i+2], color=plt.cm.jet(cn(z[i])))

plt.show()
```
